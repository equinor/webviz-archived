```python
import pandas as pd
import numpy as np
from webviz.page_elements import Histogram, BarChart, LineChart, FanChart, HeatMap, TornadoPlot
```

```python
# Histogram example:

normal = [x for x in np.random.normal(size=1000).tolist()]
poisson = [x for x in np.random.poisson(10, 1000).tolist()]
triangular = [x for x in np.random.triangular(0, 10, 20, 1000).tolist()]

df_histogram = pd.DataFrame({'normal': normal, 'poisson': poisson,
                    'triangular': triangular})

Histogram(df_histogram)
```

```python
# BarChart example:

bars1 = [10, 15, 13, 17]
bars2 = [16, 5, 11, 9]

bars = pd.DataFrame({'bars1': bars1, 'bars2': bars2})

BarChart(bars)
```

```python
# LineChart example:

from pandas.compat import StringIO

df = pd.read_csv(StringIO("""index,line1,line2,category,dateslider
2012-01-01,1,3,A,2012-01-01
2012-01-02,20,4,B,2012-01-01
2012-01-03,3,1,A,2012-01-01
2012-01-04,5,10,B,2012-01-01
2012-01-01,1,3,A,2012-01-01
2012-01-02,20,4,B,2012-01-01
2012-01-03,3,1,A,2012-01-01
2012-01-04,5,10,B,2012-01-01
2012-01-01,1,1,A,2012-01-01
2012-01-02,5,4,B,2012-01-01
2012-01-05,6,10,A,2012-01-02
2012-01-06,5,2,B,2012-01-02
2012-01-01,7,14,A,2012-01-02
2012-01-03,5,2,B,2012-01-02
2012-01-04,4,6,A,2012-01-02"""))

LineChart(df, logy=True, slider_columns=['dateslider'], dropdown_columns=['category'])
```

```python
# FanChart example:

index = ['2012-01-01', '2012-01-02', '2012-01-03', '2012-01-04']
name = ['line-1', 'line-1', 'line-1', 'line-1']

mean = [10, 15, 13, 17]
p10 = [11, 16, 13, 18]
p90 = [9, 14, 12, 16]
areaMax = [16, 17, 16, 19]
areaMin = [4, 1, 9, 8]

lines = pd.DataFrame({
  'index': index,
  'name': name,
  'mean': mean,
  'p10': p10,
  'p90': p90,
  'max': areaMax,
  'min': areaMin
})

observations = pd.DataFrame({
  'name': ['line-2', 'line-3'],
  'value': [4, 3],
  'error': 2
})

FanChart(lines, observations)
```

```python
# HeatMap example:

data = pd.DataFrame(
        [[1, 20, 30, 50, 1], [20, 1, 60, 80, 30], [30, 60, 1, -10, 20]],
        columns=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
        index=['Morning', 'Afternoon', 'Evening']
        )

HeatMap(data)
```

```python
# TornadoPlot example:

high = [0.8, 1, 0.3, 0.4]
low = [0.5, -0.7, -.5, -0.1]
index = ['A', 'B', 'C', 'D']

bars = pd.DataFrame(
    {'low': low, 'high': high},
    index=index
)

TornadoPlot(bars)
```
